```python
from diffusers import StableDiffusionXLPipeline
import torch

pipe = StableDiffusionXLPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, variant="fp16", use_safetensors=True
)
pipe.to("cuda")

```

```python
keyword = []
races = ["white","black","asian","indian"]
genders = ["man","woman"]
ages = ["child", "young adult", "middle-aged adult", "old"]
for race in races:
  for gender in genders:
    for age in ages:
      keyword.append(f"{race}, {gender}, {age}, ")
```

```python
# prompt = "a carved into a travertine terrace, elegant gleaming intricate contraption pendant on ornate leather, cinematic lighting, dramatic atmosphere, detailed and intricate environment, art by wlop, mars ravelo, greg rutowski, artstation"
prompt = "human face, high details, 4k, photograph, far portrait, one person"
count = 1
# image here is in [PIL format]
for i in keyword:
  for j in range(30):
    image = pipe(f"{i}{prompt}").images[0]
    image.save(f"/content/drive/MyDrive/crawl_stable_diffusion/stable_diffusion/image_{count}.png")
    count = count + 1
image
```
